**Problem statement:**

Spam text message classification is a crucial task in natural language processing (NLP) due to the pervasive nature of unsolicited and harmful messages, which can lead to user frustration, loss of productivity, and potential security risks. The presence of noise in the dataset, such as spelling mistakes, grammatical errors, and irrelevant information, poses significant challenges in accurately classifying these messages. Effective noise handling is essential to improve the performance and reliability of text classification models.

**Objective:**

The objective is to develop a robust text cleaning and processing pipeline for Spam Text Message Classification. This involves implementing noise detection, removal, and reduction techniques using NLTK and SKlearn packages to enhance data quality and consistency.

**Dataset:**

The dataset consists of text messages labeled as either "ham" (non-spam) or "spam". Each entry in the dataset includes a message and its corresponding category.

The dataset is structured as follows:

* Category: Indicates whether the message is "ham" (non-spam) or "spam".
* Message: Contains the text content of the message.






```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import string as st
import re
import nltk
from nltk import PorterStemmer, WordNetLemmatizer
import matplotlib.pyplot as plt
```

```python
# Read the data. Here it is already in .csv format.
data = pd.read_csv('/content/SPAM text message data.csv')
data.head()
```

```python
data.shape
```

## **Text cleaning and processing steps**



><image src="https://miro.medium.com/v2/resize:fit:828/format:webp/0*Fx8WJk9fH4Rf2qiQ.jpg">

* Remove punctuations

* Convert text to tokens

* Remove tokens of length less than or equal to 3

* Remove stopwords using NLTK corpus

* stopwords list to match

* Apply stemming

* Apply lemmatization

* Convert words to feature vectors

```python
# Remove all punctuations from the text

def remove_punct(text):
    return ("".join([ch for ch in text if ch not in st.punctuation]))
```

```python
data['removed_punc'] = data['Message'].apply(lambda x: remove_punct(x))
data.head()
```

Now, Convert text to lower case tokens. Here, split() is applied on white-spaces. But, it could be applied on special characters, tabs or any other string based on which text is to be seperated into tokens.


```python
# lowecase text
def tokenize(text):
    text = re.split('\s+' ,text)
    return [x.lower() for x in text]
```

```python
data['tokens'] = data['removed_punc'].apply(lambda msg : tokenize(msg))
data.head()
```

```python
# Remove tokens of length less than 3
def remove_small_words(text):
    return [x for x in text if len(x) > 3 ]
```

```python
data['larger_tokens'] = data['tokens'].apply(lambda x : remove_small_words(x))
data.head()
```

**Remove stopwords:**

Here, NLTK corpus list is used for a match. However, a customized user-defined list could be created and used to limit the matches in input text.

```python
nltk.download('stopwords')
def remove_stopwords(text):
    return [word for word in text if word not in nltk.corpus.stopwords.words('english')]
```

```python
data['clean_tokens'] = data['larger_tokens'].apply(lambda x : remove_stopwords(x))
data.head()
```

**Stemming:**

Apply stemming to convert tokens to their root form. This is a rule-based process of word form conversion where word-suffixes are truncated irrespective of whether the root word is an actual word in the language dictionary.

```python
# Apply stemming to get root words
def stemming(text):
    ps = PorterStemmer()
    return [ps.stem(word) for word in text]
```

```python
data['stem_words'] = data['clean_tokens'].apply(lambda wrd: stemming(wrd))
data.head()
```

**Lemmatization:**

Lemmatization converts word to it's dictionary base form. This process takes language grammar and vocabulary into consideration while conversion. Hence, it is different from Stemming in that it does not merely truncate the suffixes to get the root word.

```python
# Apply lemmatization on tokens
nltk.download('wordnet')
def lemmatize(text):
    word_net = WordNetLemmatizer()
    return [word_net.lemmatize(word) for word in text]
```

```python
data['lemma_words'] = data['clean_tokens'].apply(lambda x : lemmatize(x))
data.head()
```

```python
# Create sentences to get clean text as input for vectors

def return_sentences(tokens):
    return " ".join([word for word in tokens])
```

```python
data['clean_text'] = data['lemma_words'].apply(lambda x : return_sentences(x))
data.head()
```

**Wordcloud**

```python
# Generate a basic word cloud
from wordcloud import WordCloud, ImageColorGenerator

text = " ".join([x for x in data['clean_text']])
# Create and generate a word cloud image:
wordcloud = WordCloud(max_font_size=30, max_words=1000).generate(text)

# Display the generated image:
plt.figure(figsize= [20,10])
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# Prepare data for the model. Convert label/category in to binary

data['Category'] = [1 if x == 'spam' else 0 for x in data['Category']]
data.head()
```

All the approaches above helped in reducing the disruptive effect of noise. Now, the data is cleaned and can be further used for classification process.

```python
# Split data in to training, testing sets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

text_train, text_test, y_train, y_test = train_test_split(data['clean_text'], data['Category'], test_size = 0.2)
print(text_train.shape)
print(text_test.shape)
```

**TF-IDF : Term Frequency - Inverse Document Frequency**

The term frequency is the number of times a term occurs in a document. Inverse document frequency is an inverse function of the number of documents in which that a given word occurs.

The product of these two terms gives tf-idf weight for a word in the corpus. The higher the frequency of occurrence of a word, lower is it's weight and vice-versa. This gives more weightage to rare terms in the corpus and penalizes more commonly occuring terms.
Other widely used vectorizer is Count vectorizer which only considers the frequency of occurrence of a word across the corpus.

```python
# Convert lemmatized words to Tf-Idf feature vectors

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_train = tfidf.fit_transform(text_train)
X_test = tfidf.transform(text_test)
print(X_train.toarray())
print(X_train.shape)
print(X_test.toarray())
print(X_test.shape)
```

```python
# Get a few feature names in the vector
tfidf.get_feature_names_out()[-5::]
```

```python
# 1st Model - Logistic Regression

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter = 500)
lr.fit(X_train, y_train)
print('Logistic Regression model fitted..')

pred = lr.predict(X_test)
print("Accuracy score : {}".format(accuracy_score(y_test, pred)))
print("Confusion matrix : \n {}".format(confusion_matrix(y_test, pred)))
```

```python
# 2nd Model - Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

print('Random Forest Classifier model fitted..')
pred = rf.predict(X_test)
print("Accuracy score : {}".format(accuracy_score(y_test, pred)))
print("Confusion matrix : \n {}".format(confusion_matrix(y_test, pred)))
```

**Inference:**

Based on the accuracy results and confusion matrix comparisons, the Random Forest Classifier outperformed the Logistic Regression model in both metrics. The Random Forest achieved a higher accuracy of 97.4% compared to Logistic Regression's 95.1%. In terms of the confusion matrix, the Random Forest had fewer misclassifications overall, with 0 false positives and 29 false negatives, indicating strong performance in correctly predicting both classes. In contrast, the Logistic Regression model, while still performing well with 94 true positives and 966 true negatives, had more false negatives (53) and a small number of false positives (2). This suggests that the Random Forest Classifier's ensemble approach effectively handled the dataset's complexities, capturing more nuanced relationships between features and outcomes compared to the linear assumptions of Logistic Regression, thereby yielding superior predictive accuracy and robustness.
